### Perdidas en cosecha

```python
# Cargamos el archivo Excel - RFF Procesado
# --------------

ruta_path = 'C:/cesar_quezada/ProyectosIA/Proyectos_PALMAS'               # -> Casa
# ruta_path = 'C:/Users/usuario/Documents/DOCUMENTOS_CONSULTOR_5_BI'      # -> Trabajo

path_perdida_cosecha_transporte_puesto_pw = '/PROYECTO_01_analitica_descriptiva_TEA/data/external/bbdd_perdidas_cosecha/PERDIDA PALMAWASI PLANTACION ATC.xlsx'
path_perdida_cosecha_transporte_bin_pw = '/PROYECTO_01_analitica_descriptiva_TEA/data/external/bbdd_perdidas_cosecha/PERDIDA PALMAWASI PLANTACION ATC.xlsx'
path_perdida_cosecha_puesto_pw = '/PROYECTO_01_analitica_descriptiva_TEA/data/external/bbdd_perdidas_cosecha/PERDIDA PALMAWASI PLANTACION ATC.xlsx'

path_perdida_cosecha_transporte_puesto_sh = '/PROYECTO_01_analitica_descriptiva_TEA/data/external/bbdd_perdidas_cosecha/PERDIDA SHANUSI PLANTACION ATC.xlsx'
path_perdida_cosecha_transporte_bin_sh = '/PROYECTO_01_analitica_descriptiva_TEA/data/external/bbdd_perdidas_cosecha/PERDIDA SHANUSI PLANTACION ATC.xlsx'
path_perdida_cosecha_puesto_sh = '/PROYECTO_01_analitica_descriptiva_TEA/data/external/bbdd_perdidas_cosecha/PERDIDA SHANUSI PLANTACION ATC.xlsx'




# Cargar la hoja 2 (indexada desde 0, por lo tanto la hoja 2 es 'sheet_name=1')
df_perdida_cosecha_transporte_puesto_pw = pd.read_excel(ruta_path + path_perdida_cosecha_transporte_puesto_pw,sheet_name='TRANSPORTE PUESTO')
# Eliminar la columna 'columna_a_eliminar'
df_perdida_cosecha_transporte_puesto_pw = df_perdida_cosecha_transporte_puesto_pw.drop(columns=['% PERDIDA'])
df_perdida_cosecha_transporte_puesto_pw.columns = ['FECHA','SECTOR','COSECHA TM','PERDIDA TM']
# -
df_perdida_cosecha_transporte_bin_pw = pd.read_excel(ruta_path + path_perdida_cosecha_transporte_bin_pw,sheet_name='TRANSPORTE BIN')
df_perdida_cosecha_transporte_bin_pw.columns = ['FECHA','SECTOR','COSECHA TM','PERDIDA TM']
# -
df_perdida_cosecha_puesto_pw = pd.read_excel(ruta_path + path_perdida_cosecha_puesto_pw,sheet_name='COSECHA PUESTO')
df_perdida_cosecha_puesto_pw.columns = ['FECHA','SECTOR','COSECHA TM','PERDIDA TM']

df_perdida_cosecha_transporte_puesto_sh = pd.read_excel(ruta_path + path_perdida_cosecha_transporte_puesto_sh,sheet_name='TRANSPORTE PUESTO')
df_perdida_cosecha_transporte_puesto_sh.columns = ['FECHA','SECTOR','COSECHA TM','PERDIDA TM']
# -
df_perdida_cosecha_transporte_bin_sh = pd.read_excel(ruta_path + path_perdida_cosecha_transporte_bin_sh,sheet_name='TRANSPORTE BIN')
df_perdida_cosecha_transporte_bin_sh.columns = ['FECHA','SECTOR','COSECHA TM','PERDIDA TM']
# -
df_perdida_cosecha_puesto_sh = pd.read_excel(ruta_path + path_perdida_cosecha_puesto_sh,sheet_name='COSECHA PUESTO')
df_perdida_cosecha_puesto_sh.columns = ['FECHA','SECTOR','COSECHA TM','PERDIDA TM']

# AÃ±adiendo los tipos de perdidas
# ----------
df_perdida_cosecha_transporte_puesto_pw['TIPO_PERDIDA'] = 'Transporte puesto'
df_perdida_cosecha_transporte_bin_pw['TIPO_PERDIDA'] = 'Transporte bin'
df_perdida_cosecha_puesto_pw['TIPO_PERDIDA'] = 'Cosecha puesto'
# ----------
df_perdida_cosecha_transporte_puesto_sh['TIPO_PERDIDA'] = 'Transporte puesto'
df_perdida_cosecha_transporte_bin_sh['TIPO_PERDIDA'] = 'Transporte bin'
df_perdida_cosecha_puesto_sh['TIPO_PERDIDA'] = 'Cosecha puesto'

```

Transformando datos de Palmawasi y Nuevo Horizonte

```python
# Concatenar los DataFrames
df_perdida_otros_pw_nh = pd.concat([df_perdida_cosecha_transporte_puesto_pw, 
                                    df_perdida_cosecha_transporte_bin_pw, 
                                    df_perdida_cosecha_puesto_pw], ignore_index=True)

df_perdida_otros_pw_nh['PLANTACION'] = np.where(df_perdida_otros_pw_nh['SECTOR'] == 'SE05', 'Nuevo Horizonte', 'Palmawasi')

df_perdida_otros_pw_nh.head()
```

Transformando datos de Shanuzi

```python
# Concatenar los DataFrames
df_perdida_otros_sh = pd.concat([df_perdida_cosecha_transporte_puesto_sh,
                                 df_perdida_cosecha_transporte_bin_sh, 
                                 df_perdida_cosecha_puesto_sh], ignore_index=True)

df_perdida_otros_sh['PLANTACION'] = 'Shanuzi'

df_perdida_otros_sh.head()
```

Uniendo los dataframe

```python
df_perdida_otros_sh.columns == df_perdida_otros_pw_nh.columns
```

```python
df_perdida_otros_full = pd.concat([df_perdida_otros_pw_nh,
                                   df_perdida_otros_sh],
                                   ignore_index= True)


df_perdida_otros_full.head()
```

```python
df_perdida_otros_full.dtypes
```

```python
df_perdida_otros_full_group = df_perdida_otros_full.groupby(['FECHA','PLANTACION','TIPO_PERDIDA']).agg({'COSECHA TM':'sum',
                                                                                                        'PERDIDA TM':'sum'}).reset_index()
df_perdida_otros_full_group.head()
```

#### Guardamos los datos

```python
# Guardamos la data en processed
# -----
# Especifica la ruta donde deseas guardar el CSV
ruta_perdida_otros_full_group =  '/PROYECTO_01_analitica_descriptiva_TEA/data/processed/df_perdida_otros_full_group.csv'  # Cambia esta ruta a la que necesites
# Guardar el DataFrame en formato CSV
df_perdida_otros_full_group.to_csv(ruta_path + ruta_perdida_otros_full_group, index=False)
print(f'Datos guardados en {ruta_path + ruta_perdida_otros_full_group}')

df_perdida_otros_full_group.head(10)
```
