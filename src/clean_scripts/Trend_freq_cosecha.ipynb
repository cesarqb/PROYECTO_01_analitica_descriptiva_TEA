```python
#from google.colab import files
import pandas as pd
import numpy as np
import calendar
# import warnings

import matplotlib.pyplot as plt
```

```python
# Cargamos el archivo Excel - RFF Procesado
# --------------

ruta_path = 'C:/cesar_quezada/ProyectosIA/Proyectos_PALMAS'               # -> Casa
# ruta_path = 'C:/Users/usuario/Documents/DOCUMENTOS_CONSULTOR_5_BI'      # -> Trabajo

path_freq_cosecha_pw_nh = '/PROYECTO_001_analitica_descriptiva_TEA/data/external/bbdd_frecuencia_cosecha/FRECUENCIA DE COSECHA POR SECTORES.xlsx'
path_freq_cosecha_sh = '/PROYECTO_001_analitica_descriptiva_TEA/data/external/bbdd_frecuencia_cosecha/FRECUENCIA DE COSECHA POR SECTORES.xlsx'

# Cargar la hoja 2 (indexada desde 0, por lo tanto la hoja 2 es 'sheet_name=1')
df_freq_cosecha_pw_nh = pd.read_excel(ruta_path + path_freq_cosecha_pw_nh, sheet_name='PW_T')
df_freq_cosecha_sh = pd.read_excel(ruta_path + path_freq_cosecha_pw_nh, sheet_name='SH_T')
```

```python
# Crear un diccionario de mapeo
mapping = {
    'SE01': 'Palmawasi',
    'SE02': 'Palmawasi',
    'sector3': 'Palmawasi',
    'SE04': 'Palmawasi',
    'SE4F': 'Palmawasi',
    'SE05': 'Nuevo Horizonte'
}
# Usar map y fillna
df_freq_cosecha_pw_nh['PLANTACION'] = df_freq_cosecha_pw_nh['SECTOR'].map(mapping).fillna('Desconocido')
# --
df_freq_cosecha_sh['PLANTACION'] = 'Shanuzi'
```

```python
df_freq_cosecha_pw_nh.head()
```

```python
df_freq_cosecha_sh.head()
```

#### Generando base total - Frecuencia cosecha

```python
df_freq_cosecha_concat = pd.concat([df_freq_cosecha_pw_nh,
                                    df_freq_cosecha_sh],
                                    axis=0,
                                    ignore_index=True)

print(df_freq_cosecha_concat.dtypes)
df_freq_cosecha_concat.head()
```

```python
df_freq_cosecha_full = df_freq_cosecha_concat.groupby(['FECHA','PLANTACION']).agg({'FRECUENCIA':'mean'}).reset_index()
df_freq_cosecha_full.head()
```

```python
# Especifica la ruta donde deseas guardar el CSV
ruta_freq_cosecha_full =  '/PROYECTO_001_analitica_descriptiva_TEA/data/processed/df_freq_cosecha_full.csv'  # Cambia esta ruta a la que necesites
# Guardar el DataFrame en formato CSV
df_freq_cosecha_full.to_csv(ruta_path + ruta_freq_cosecha_full, index=False)
print(f'Datos guardados en {ruta_path + ruta_freq_cosecha_full}')

df_freq_cosecha_full.head()
```
