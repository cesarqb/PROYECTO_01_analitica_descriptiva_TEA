### Cargando datos de calida racimos propios

```python
# Cargamos el archivo Excel - RFF Procesado
# --------------

ruta_path = 'C:/cesar_quezada/ProyectosIA/Proyectos_PALMAS'               # -> Casa
# ruta_path = 'C:/Users/usuario/Documents/DOCUMENTOS_CONSULTOR_5_BI'      # -> Trabajo

path_calidad_racimos_propios_pw_nh = '/PROYECTO_01_analitica_descriptiva_TEA/data/external/bbdd_calidad_racimos_propios/CALIDAD PROPIOS SECTORIZADA PW.xlsx'
path_calidad_racimos_propios_sh = '/PROYECTO_01_analitica_descriptiva_TEA/data/external/bbdd_calidad_racimos_propios/CALIDAD PROPIOS SECTORIZADA SH.xlsx'

# Cargar la hoja 2 (indexada desde 0, por lo tanto la hoja 2 es 'sheet_name=1')
df_calidad_racimos_propios_pw_nh = pd.read_excel(ruta_path + path_calidad_racimos_propios_pw_nh,sheet_name='sheet1')
df_calidad_racimos_propios_sh = pd.read_excel(ruta_path + path_calidad_racimos_propios_sh,sheet_name='sheet1')
```

#### Calidad racimos propios Palmawasi y Nuevo Horizonte

```python
df_calidad_racimos_propios_pw_nh.head()
```

```python
# Creando variable FECHA
df_calidad_racimos_propios_pw_nh['FECHA'] = df_calidad_racimos_propios_pw_nh['DT_FECHA_EVALUACION']

# Crear un diccionario de mapeo
mapping = {
    'I': 'SE01',
    'I ': 'SE01',
    'II': 'SE02',
    'III': 'SE03',
    'IV': 'SE04',
    'IV-F': 'SE4F',
    'V': 'SE05',
    'VI': 'Desconocido'
}
# Creando variable SECTOR
df_calidad_racimos_propios_pw_nh['SECTOR'] = df_calidad_racimos_propios_pw_nh['VCH_SECTOR'].map(mapping).fillna('Desconocido')

# Creando variable PLANTACION
df_calidad_racimos_propios_pw_nh['PLANTACION'] = np.where(df_calidad_racimos_propios_pw_nh['SECTOR'] == 'SE05', 'Nuevo Horizonte', 'Palmawasi')

df_calidad_racimos_propios_pw_nh.head()
```

#### Enriqueciendo base datos - Calidad Racimos propios

```python
# añadiendo columna de pesos por sector - BD antiguedad_fruta_total_pesos
# ####################
# ####################
# ####################

ruta_path = 'C:/cesar_quezada/ProyectosIA/Proyectos_PALMAS'               # -> Casa
# ruta_path = 'C:/Users/usuario/Documents/DOCUMENTOS_CONSULTOR_5_BI'      # -> Trabajo

ruta_path_fruta_total_pesos = '/PROYECTO_01_analitica_descriptiva_TEA/data/processed/df_antiguedad_fruta_total_pesos_group.csv'
df_antiguedad_fruta_total_pesos_group = pd.read_csv(ruta_path + ruta_path_fruta_total_pesos)
df_antiguedad_fruta_total_pesos_group.tail()
```

```python
df_antiguedad_fruta_total_pesos_group.groupby(['PLANTACION','SECTOR']).size()
```

```python
# Validando que no haya duplicados
# -----
df_antiguedad_fruta_total_pesos_group['FECHA'] = pd.to_datetime(df_antiguedad_fruta_total_pesos_group['FECHA'])
df_sin_duplicados = df_antiguedad_fruta_total_pesos_group[['FECHA','PLANTACION','SECTOR']].drop_duplicates()
print(df_antiguedad_fruta_total_pesos_group.shape)
print(df_sin_duplicados.shape)
```

```python
#df_antiguedad_fruta_total_pesos_group.groupby('PLANTACION').size()
df_antiguedad_fruta_total_pesos_group.groupby(['PLANTACION','SECTOR']).agg({'FECHA':['min','max']}).reset_index()
```

```python
# Validando base: calidad_racimos_propios_total
# ####################
# ####################
# ####################

# df_calidad_racimos_propios_total.groupby('PLANTACION').size()
df_calidad_racimos_propios_total.groupby(['PLANTACION','SECTOR']).agg({'FECHA':['min','max']}).reset_index()
```

```python
# Validando que no haya duplicados
# -----
df_sin_duplicados_calidad_racimos = df_calidad_racimos_propios_total[['FECHA','PLANTACION','SECTOR']].drop_duplicates()
print(df_sin_duplicados_calidad_racimos.shape)
print(df_calidad_racimos_propios_total.shape)
```

```python
## Agrupando por PLANTACION
## -----
df_calidad_racimos_propios_total.head()
```

```python
df_calidad_racimos_propios_total_agrupado = df_calidad_racimos_propios_total.groupby(['FECHA','PLANTACION']).agg({'VERDES':'sum',
                                                                                                                  'MADUROS':'sum',
                                                                                                                  'SMADUROS':'sum',
                                                                                                                  'PEDUNCULO':'sum',
                                                                                                                  'ESCOBAJO':'sum',
                                                                                                                  'MALFORMADOA':'sum',
                                                                                                                  'MALFORMADOB':'sum'}).reset_index()

df_calidad_racimos_propios_total_agrupado.head()
```

```python
df_antiguedad_fruta_total_pesos_group.head()
```

```python
df_antiguedad_fruta_total_pesos_group_agrupado = df_antiguedad_fruta_total_pesos_group.groupby(['FECHA','PLANTACION']).agg({'Peso':'sum',
                                                                                                                            'Racimos':'sum'}).reset_index()
df_antiguedad_fruta_total_pesos_group_agrupado.head()
```

```python
df_antiguedad_fruta_total_pesos_group_agrupado
```

```python
## Añadiendo Pesos a BBDD calidad racimos propios
## ------------------
## ------------------

# Realiza la unión de los DataFrames usando las columnas comunes
df_calidad_racimos_propios_total_full = pd.merge(df_calidad_racimos_propios_total_agrupado, 
                                                 df_antiguedad_fruta_total_pesos_group_agrupado, 
                                                 on=['FECHA', 'PLANTACION'], how='outer')

df_calidad_racimos_propios_total_full.head()
```

```python
# Viendo resumenes
# ----------

df_calidad_racimos_propios_total_full.tail()
```

```python
# Copiamos el datframe
# -----

# Especifica la ruta donde deseas guardar el CSV
ruta_calidad_racimos_propios_total_full =  '/PROYECTO_01_analitica_descriptiva_TEA/data/processed/df_calidad_racimos_propios_total_full.csv'  # Cambia esta ruta a la que necesites
df_calidad_racimos_propios_total_full.to_csv(ruta_path + ruta_calidad_racimos_propios_total_full, index=False)
print(f'Datos guardados en {ruta_path + ruta_calidad_racimos_propios_total_full}')

df_calidad_racimos_propios_total_full.head()
```

### Calidad racimos Terceros

```python
# Cargamos el archivo Excel - Calidad racimos
# --------------

ruta_path = 'C:/cesar_quezada/ProyectosIA/Proyectos_PALMAS'               # -> Casa
# ruta_path = 'C:/Users/usuario/Documents/DOCUMENTOS_CONSULTOR_5_BI'      # -> Trabajo

path_calidad_racimos_terceros_pw = '/PROYECTO_01_analitica_descriptiva_TEA/data/external/bbdd_calidad_racimos_terceros/c7fe214a-8a22-458d-abf3-PW.xlsx'
path_calidad_racimos_terceros_nh = '/PROYECTO_01_analitica_descriptiva_TEA/data/external/bbdd_calidad_racimos_terceros/2bd017e5-a016-4eb5-b512-NH.xlsx'
path_calidad_racimos_terceros_sh = '/PROYECTO_01_analitica_descriptiva_TEA/data/external/bbdd_calidad_racimos_terceros/7486d7d0-d18a-4276-a7a0-SH.xlsx'
path_calidad_racimos_terceros_tl = '/PROYECTO_01_analitica_descriptiva_TEA/data/external/bbdd_calidad_racimos_terceros/b381e2e3-b5fa-4615-90cb-TL.xlsx'


# Cargar la hoja 2 (indexada desde 0, por lo tanto la hoja 2 es 'sheet_name=1')
df_calidad_racimos_terceros_pw = pd.read_excel(ruta_path + path_calidad_racimos_terceros_pw,sheet_name='Sheet1')
df_calidad_racimos_terceros_nh = pd.read_excel(ruta_path + path_calidad_racimos_terceros_nh,sheet_name='Sheet1')
df_calidad_racimos_terceros_sh = pd.read_excel(ruta_path + path_calidad_racimos_terceros_sh,sheet_name='Sheet1')
df_calidad_racimos_terceros_tl = pd.read_excel(ruta_path + path_calidad_racimos_terceros_tl,sheet_name='Sheet1')

# Eliminar las filas donde cualquiera de las columnas tenga el valor 'Totales' o NaN
df_calidad_racimos_terceros_pw = df_calidad_racimos_terceros_pw[~df_calidad_racimos_terceros_pw[['AÑO', 'MES']].isin(['Totales']).any(axis=1) & df_calidad_racimos_terceros_pw[['AÑO', 'MES']].notna().all(axis=1)]
df_calidad_racimos_terceros_nh = df_calidad_racimos_terceros_nh[~df_calidad_racimos_terceros_nh[['AÑO', 'MES']].isin(['Totales']).any(axis=1) & df_calidad_racimos_terceros_nh[['AÑO', 'MES']].notna().all(axis=1)]
df_calidad_racimos_terceros_sh = df_calidad_racimos_terceros_sh[~df_calidad_racimos_terceros_sh[['AÑO', 'MES']].isin(['Totales']).any(axis=1) & df_calidad_racimos_terceros_sh[['AÑO', 'MES']].notna().all(axis=1)]
df_calidad_racimos_terceros_tl = df_calidad_racimos_terceros_tl[~df_calidad_racimos_terceros_tl[['AÑO', 'MES']].isin(['Totales']).any(axis=1) & df_calidad_racimos_terceros_tl[['AÑO', 'MES']].notna().all(axis=1)]


df_calidad_racimos_terceros_pw['PLANTACION'] = 'Palmawasi'
df_calidad_racimos_terceros_nh['PLANTACION'] = 'Nuevo Horizonte'
df_calidad_racimos_terceros_sh['PLANTACION'] = 'Shanuzi'
df_calidad_racimos_terceros_tl['PLANTACION'] = 'Tulumayo'

# Concatenar los DataFrames
df_sabana_calidad_terceros_full = pd.concat([df_calidad_racimos_terceros_pw, df_calidad_racimos_terceros_nh,df_calidad_racimos_terceros_sh,df_calidad_racimos_terceros_tl], ignore_index=True)

print(df_sabana_calidad_terceros_full.dtypes)
df_sabana_calidad_terceros_full.head(10)
```
