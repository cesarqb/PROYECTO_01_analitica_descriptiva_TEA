## Transporte en balanza en el acopio (Propios y Terceros)

```python
# Cargamos el archivo Excel - RFF Procesado
# --------------

ruta_path = 'C:/cesar_quezada/ProyectosIA/Proyectos_PALMAS'               # -> Casa
# ruta_path = 'C:/Users/usuario/Documents/DOCUMENTOS_CONSULTOR_5_BI'      # -> Trabajo

path_sabana_participacion_nh = '/PROYECTO_01_analitica_descriptiva_TEA/data/external/bbdd_participacion/f431f786-ea60-42d1-participacion-nh.xlsx'
path_sabana_participacion_pw = '/PROYECTO_01_analitica_descriptiva_TEA/data/external/bbdd_participacion/bcd00108-c863-45af-participacion-pw.xlsx'
path_sabana_participacion_sh = '/PROYECTO_01_analitica_descriptiva_TEA/data/external/bbdd_participacion/60af1148-0ee5-4d21-participacion-sh.xlsx'
path_sabana_participacion_tl = '/PROYECTO_01_analitica_descriptiva_TEA/data/external/bbdd_participacion/e8fdb299-c22e-45c1-participacion-tl.xlsx'


# Cargar la hoja 2 (indexada desde 0, por lo tanto la hoja 2 es 'sheet_name=1')
df_sabana_participacion_nh = pd.read_excel(ruta_path + path_sabana_participacion_nh,sheet_name='Sheet1')
df_sabana_participacion_pw = pd.read_excel(ruta_path + path_sabana_participacion_pw,sheet_name='Sheet1')
df_sabana_participacion_sh = pd.read_excel(ruta_path + path_sabana_participacion_sh,sheet_name='Sheet1')
df_sabana_participacion_tl = pd.read_excel(ruta_path + path_sabana_participacion_tl,sheet_name='Sheet1')

# Eliminar las filas donde cualquiera de las columnas tenga el valor 'Totales' o NaN
df_sabana_participacion_nh = df_sabana_participacion_nh[~df_sabana_participacion_nh[["PTNOMBREMES", "Date(PTDIA, 'DD/MM/YYYY')"]].isin(['Totales']).any(axis=1) & df_sabana_participacion_nh[["PTNOMBREMES", "Date(PTDIA, 'DD/MM/YYYY')"]].notna().all(axis=1)]
df_sabana_participacion_pw = df_sabana_participacion_pw[~df_sabana_participacion_pw[["PTNOMBREMES", "Date(PTDIA, 'DD/MM/YYYY')"]].isin(['Totales']).any(axis=1) & df_sabana_participacion_pw[["PTNOMBREMES", "Date(PTDIA, 'DD/MM/YYYY')"]].notna().all(axis=1)]
df_sabana_participacion_sh = df_sabana_participacion_sh[~df_sabana_participacion_sh[["PTNOMBREMES", "Date(PTDIA, 'DD/MM/YYYY')"]].isin(['Totales']).any(axis=1) & df_sabana_participacion_sh[["PTNOMBREMES", "Date(PTDIA, 'DD/MM/YYYY')"]].notna().all(axis=1)]
df_sabana_participacion_tl = df_sabana_participacion_tl[~df_sabana_participacion_tl[["PTNOMBREMES", "Date(PTDIA, 'DD/MM/YYYY')"]].isin(['Totales']).any(axis=1) & df_sabana_participacion_tl[["PTNOMBREMES", "Date(PTDIA, 'DD/MM/YYYY')"]].notna().all(axis=1)]

df_sabana_participacion_nh['PLANTACION'] = 'Nuevo Horizonte'
df_sabana_participacion_pw['PLANTACION'] = 'Palmawasi'
df_sabana_participacion_sh['PLANTACION'] = 'Shanuzi'
df_sabana_participacion_tl['PLANTACION'] = 'Tulumayo'

# Concatenar los DataFrames
df_sabana_participacion_full = pd.concat([df_sabana_participacion_nh, df_sabana_participacion_pw, df_sabana_participacion_sh, df_sabana_participacion_tl], ignore_index=True)

# Creando variable FECHA
df_sabana_participacion_full['FECHA'] = pd.to_datetime(df_sabana_participacion_full["Date(PTDIA, 'DD/MM/YYYY')"])

print(df_sabana_participacion_full.dtypes)

# Guardamos la data en processed
# -----
# Especifica la ruta donde deseas guardar el CSV
ruta_sabana_participacion_full =  '/PROYECTO_01_analitica_descriptiva_TEA/data/processed/df_sabana_participacion_full.csv'  # Cambia esta ruta a la que necesites
# Guardar el DataFrame en formato CSV
df_sabana_participacion_full.to_csv(ruta_path + ruta_sabana_participacion_full, index=False)
print(f'Datos guardados en {ruta_path + ruta_sabana_participacion_full}')

df_sabana_participacion_full.head(10)
```
