```python
#from google.colab import files
import pandas as pd
import numpy as np
import calendar
# import warnings

import matplotlib.pyplot as plt
```

### Cargando datos de antiguedad de frutos

```python
# Cargamos el archivo Excel - RFF Procesado
# --------------

ruta_path = 'C:/cesar_quezada/ProyectosIA/Proyectos_PALMAS'               # -> Casa
# ruta_path = 'C:/Users/usuario/Documents/DOCUMENTOS_CONSULTOR_5_BI'      # -> Trabajo

path_antiguedad_fruta_pw_nh = '/PROYECTO_01_analitica_descriptiva_TEA/data/external/bbdd_antiguedad_plantas/AÑO RACIMOS PESO PALMAWASI 1.xlsx'
path_antiguedad_fruta_sh = '/PROYECTO_01_analitica_descriptiva_TEA/data/external/bbdd_antiguedad_plantas/AÑO RACIMOS PESO SHANUSI 1.xlsx'


# Cargar la hoja 2 (indexada desde 0, por lo tanto la hoja 2 es 'sheet_name=1')
df_antiguedad_fruta_pw_nh = pd.read_excel(ruta_path + path_antiguedad_fruta_pw_nh,sheet_name='sheet1')
df_antiguedad_fruta_sh = pd.read_excel(ruta_path + path_antiguedad_fruta_sh,sheet_name='sheet1')
```

#### Antigueda Fruta Palmawasi y Nuevo Horizonte

```python
df_antiguedad_fruta_pw_nh.tail(15)
```

```python
df_antiguedad_fruta_pw_nh['pk_tiempo'] = df_antiguedad_fruta_pw_nh['pk_tiempo'].astype(str)
df_antiguedad_fruta_pw_nh['FECHA'] = pd.to_datetime(df_antiguedad_fruta_pw_nh['pk_tiempo'], format='%Y%m%d') # Parseando Fecha
df_antiguedad_fruta_pw_nh['FECHA'] = df_antiguedad_fruta_pw_nh['FECHA'].dt.strftime('%Y-%m-%d') # Formateando Fecha

# Creando variable PLANTACION
df_antiguedad_fruta_pw_nh['PLANTACION'] = np.where(df_antiguedad_fruta_pw_nh['vch_sector'] == 'SE05', 'Nuevo Horizonte', 'Palmawasi')

print(df_antiguedad_fruta_pw_nh.groupby('PLANTACION').size())
df_antiguedad_fruta_pw_nh.tail()
```

#### Antigueda Fruta Shanusi

```python
df_antiguedad_fruta_sh.tail()
```

```python
df_antiguedad_fruta_sh['pk_tiempo'] = df_antiguedad_fruta_sh['pk_tiempo'].astype(str)
df_antiguedad_fruta_sh['FECHA'] = pd.to_datetime(df_antiguedad_fruta_sh['pk_tiempo'], format='%Y%m%d') # Parseando Fecha
df_antiguedad_fruta_sh['FECHA'] = df_antiguedad_fruta_sh['FECHA'].dt.strftime('%Y-%m-%d') # Formateando Fecha

# Creando variable PLANTACION
df_antiguedad_fruta_sh['PLANTACION'] = 'Shanuzi'

df_antiguedad_fruta_sh.head()
```

```python
df_antiguedad_fruta_sh.tail()
```

#### Generando base total - antiguedad

```python
df_antiguedad_fruta_total = pd.concat([df_antiguedad_fruta_pw_nh,df_antiguedad_fruta_sh], ignore_index=True)
print(f'dimensión de la base de datos: {df_antiguedad_fruta_total.shape}')
df_antiguedad_fruta_total.head()
```

```python
df_antiguedad_fruta_total.PLANTACION.unique()
```

```python
# Guardamos el dato de pesos de racimos x sector
# -------------------
# -------------------
df_antiguedad_fruta_total_pesos_group = df_antiguedad_fruta_total.groupby(['FECHA','PLANTACION','vch_sector']).agg({'Peso':'sum',
                                                                                                                    'Racimos':'sum'}).reset_index()
df_antiguedad_fruta_total_pesos_group = df_antiguedad_fruta_total_pesos_group.rename(columns={'vch_sector':'SECTOR'})
df_antiguedad_fruta_total_pesos_group.head()
```

```python
df_antiguedad_fruta_total_pesos_group.tail()
```

```python
df_antiguedad_fruta_total.tail()
```

```python
# Guardamos el dato de campaña de racimos x sector
# -------------------
# -------------------
df_antiguedad_fruta_total_group = df_antiguedad_fruta_total.groupby(['PLANTACION','FECHA']).agg(campania_min=('vch_campania', 'min'),
                                                                                                campania_max=('vch_campania', 'max'),
                                                                                                campania_prom=('vch_campania', 'mean'),
                                                                                                campania_median=('vch_campania', 'median'),
                                                                                                distinct_count=('vch_campania', 'nunique')).reset_index()
# Renombrar el conteo distinto
df_antiguedad_fruta_total_group.head(15)
```

```python
# Guardamos el datframe
# -----
df_antiguedad_fruta_full = df_antiguedad_fruta_total_group

# Especifica la ruta donde deseas guardar el CSV
ruta_antiguedad_fruta_full =  '/PROYECTO_01_analitica_descriptiva_TEA/data/processed/df_antiguedad_fruta_full.csv'  # Cambia esta ruta a la que necesites
df_antiguedad_fruta_full.to_csv(ruta_path + ruta_antiguedad_fruta_full, index=False)
print(f'Datos guardados en {ruta_path + ruta_antiguedad_fruta_full}')

# Especifica la ruta donde deseas guardar el CSV
ruta_peso_sector_full =  '/PROYECTO_01_analitica_descriptiva_TEA/data/processed/df_antiguedad_fruta_total_pesos_group.csv'  # Cambia esta ruta a la que necesites
df_antiguedad_fruta_total_pesos_group.to_csv(ruta_path + ruta_peso_sector_full, index=False)
print(f'Datos guardados en {ruta_path + ruta_peso_sector_full}')

df_antiguedad_fruta_full.head()
```
