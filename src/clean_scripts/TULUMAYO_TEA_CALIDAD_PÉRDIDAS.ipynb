```python
import itertools
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import LeaveOneOut

X = DF_CALIDAD_RFF_PERDIDA_SEMANAL_2024[['P.A. PROCESO (%)', 'RACIMO VERDE (%)', 'RACIMO MADURO (%)', 'PEDÚNCULO LARGO (%)', 'RACIMO SOBREMADURO (%)', 'ESCOBAJO (%)', 'RACIMO MAL FORMADO (%)']]
y = DF_CALIDAD_RFF_PERDIDA_SEMANAL_2024['TEA']

variables = X.columns
resultados = []

for L in range(1, len(variables)+1):
  for subset in itertools.combinations(variables, L):
    X_subset = X[list(subset)]

    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X_subset)

    modelo = LinearRegression()
    loo = LeaveOneOut()

    coeficientes = []
    intersecciones = []
    r2_scores = []

    for train_index, test_index in loo.split(X_scaled):
      X_train_scaled, X_test_scaled = X_scaled[train_index], X_scaled[test_index]
      y_train, y_test = y.iloc[train_index], y.iloc[test_index]

      modelo.fit(X_train_scaled, y_train)

      coeficientes.append(modelo.coef_)
      intersecciones.append(modelo.intercept_)
      r2_scores.append(modelo.score(X_train_scaled, y_train))

    coeficientes_promedio = np.round(np.mean(coeficientes, axis = 0), 3)
    interseccion_promedio = np.round(np.mean(intersecciones), 3)
    r2_promedio = np.mean(r2_scores)

    resultados.append({'variables': subset,
                       'coeficientes_promedio': coeficientes_promedio,
                       'interseccion_promedio': interseccion_promedio,
                       'r2_promedio': r2_promedio})

df_resultados = pd.DataFrame(resultados)

df_resultados['variables'] = df_resultados['variables'].apply(lambda x: ', '.join(x))
df_resultados              = df_resultados.sort_values(by = 'r2_promedio', ascending = False).reset_index(drop = True)
df_resultados_RM           = df_resultados[df_resultados['variables'].str.contains('RACIMO MADURO')].reset_index(drop = True)
df_resultados_RM
```
